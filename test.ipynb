```python
with open("test.html","r",encoding="utf-8") as f:
    soup=BeautifulSoup(f,"html.parser")
```

```python
name=soup.find_all("a")
name_list=[x.get_text(strip=True) for x in name if x != '']
```

```python
name_list=[x for x in name_list if x != '']
```

```python
print(name_list)
```

```python
price=soup.find_all("div",class_="product_price")
```

```python
price_list=[x.get_text(strip=True) for x in price]
```

```python
print(len(name_list))
print(len(price_list))
```

```python

print(price_list)
```

```python
import re
listofdigit=[]
for i in price_list:
    digits = re.findall(r"\d[\d\.]*(?=đ)", i)[0]
    digits = digits.replace(".", "") 
    listofdigit.append(digits)
print(listofdigit)
```
